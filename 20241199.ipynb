## 1. LeetCode 225. Implement Stack using Queues

````python
from queue import Queue
class MyStack:

    def __init__(self):
        self.q = Queue()

    def push(self, x: int) -> None:
        self.q.put(x)
        for _ in range(self.q.qsize()-1):
            temp = self.q.get()
            self.q.put(temp)

    def pop(self) -> int:
        return self.q.get()

    def top(self) -> int:
        t = self.q.get()
        self.q.put(t)
        for _ in range(self.q.qsize()-1):
            temp = self.q.get()
            self.q.put(temp)
        return t

    def empty(self) -> bool:
        return self.q.empty()
````

## 2. LeetCode 232. Implement Queue using Stacks

````python
class MyQueue:

    def __init__(self):
        self.s1 = []
        self.s2 = []

    def push(self, x: int) -> None:
        self.s1.append(x)

    def pop(self) -> int:
        if len(self.s2) == 0:
            while self.s1:
                temp = self.s1.pop()
                self.s2.append(temp)
        return self.s2.pop()

    def peek(self) -> int:
        if len(self.s2) == 0:
            while self.s1:
                temp = self.s1.pop()
                self.s2.append(temp)
        return self.s2[-1]

    def empty(self) -> bool:
        return len(self.s1) == 0 and len(self.s2) == 0
````

```python
from queue import Queue

def isPalindrome(str):
    if "$" not in str:
        return False

    left, right = str.split("$")
    queue1 = Queue()

    for ch in left:
        queue1.put(ch)
    for ch in reversed(right):
        if queue1.empty():
            return False
        if queue1.get() != ch:
            return False

    if queue1.empty():
        return True
    else:
        return False
```
